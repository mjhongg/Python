```python
import pandas as pd

mpg = pd.read_csv('mpg.csv')
mpg2 = mpg.copy()
```

```python
mpg2
```

```python
mpg2.assign( total = lambda x: x['cty'] + x['hwy'] )
```

```python
mpg2.assign( total = lambda x: x['cty'] + x['hwy'],
             mean = lambda x: x['total']/2 )
```

```python
mpg2.assign( total = lambda x: x['cty'] + x['hwy'],
             mean = lambda x: x['total']/2 ) \
        .sort_values('mean', ascending = False ) \
        .head(3)
```

```python
mpg.assign( total = lambda x: x['cty'] + x['hwy'],
            mean = lambda x: x['total'] / 2) \
            .sort_values('mean', ascending = False) \
            .head(3)
```
