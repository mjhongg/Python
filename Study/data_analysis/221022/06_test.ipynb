```python
import pandas as pd

midwest = pd.read_csv('midwest.csv')
midwest
```

```python
midwest = midwest.assign( popchild = (midwest['poptotal'] - midwest['popadults'])/midwest['poptotal'] * 100 )

midwest
```

```python
midwest['popchild']
```

```python
midwest.sort_values('popchild', ascending = False)[['county', 'popchild']].head(5)
```

```python
midwest
```

```python
import numpy as np

midwest = midwest.assign(grade = np.where(midwest['popchild']>=40, 'large',
                        np.where(midwest['popchild']>=30, 'middle', 'small')))

```

```python
midwest.groupby('grade').agg(n = ('county', 'count'))
```

```python
midwest.assign( asian_ratio = midwest['popasian']/midwest['poptotal']*100 )[['state', 'county', 'asian_ratio']].sort_values('asian_ratio').head(10)
```
