```python
import pandas as pd

exam = pd.read_csv('exam.csv')
exam
```

```python
exam.agg(mean_math = ('math', 'mean'))
```

```python
exam.groupby('nclass') \
    .agg(mean_math = ('math', 'mean'))
```

```python
exam.groupby('nclass', as_index = False) \
    .agg(mean_math = ('math', 'mean'))
```

```python
exam.groupby('nclass', as_index = False) \
    .agg(mean_math = ('math', 'mean'),
         sum_math = ('math', 'sum'),
         median_math = ('math', 'median'),
         n = ('math', 'count'))
```

```python
mpg = pd.read_csv('mpg.csv')
mpg
```

```python
mpg.groupby(['manufacturer', 'drv']) \
    .agg(mean_cty = ('cty', 'mean'))
```

```python
mpg.groupby(['manufacturer', 'drv']) \
    .agg(cnt = ('drv', 'count'))
```

```python
mpg.query('manufacturer == "audi"') \
    .groupby('drv') \
    .agg(n = ('drv', 'count'))
```

```python
mpg.query('manufacturer == "chevrolet"') \
    .groupby('drv') \
    .agg(n = ('drv', 'count'))
```

```python
mpg.query('category == "suv"')
```

```python
mpg.assign(total = (mpg['cty'] + mpg['hwy']) / 2) \
    .query('category == "suv"') \
    .groupby('manufacturer') \
    .agg(avg = ('total', 'mean')) \
    .sort_values('avg', ascending = False) \
    .head(5)
```

```python
mpg.query('category == "suv"') \
    .assign(total = (mpg['cty'] + mpg['hwy']) / 2) \
    .groupby('manufacturer') \
    .agg(avg = ('total', 'mean')) \
    .sort_values('avg', ascending = False) \
    .head(5)
```

```python
mpg.groupby('category') \
    .agg( avg = ('cty', 'mean'))
```

```python
mpg.groupby('category') \
    .agg( avg = ('cty', 'mean')) \
    .sort_values('avg', ascending = False)
```

```python
mpg.groupby('manufacturer').agg( mean_hwy = ('hwy', 'mean'))
```

```python
mpg.groupby('manufacturer') \
    .agg( mean_hwy = ('hwy', 'mean')) \
    .sort_values('mean_hwy', ascending=False) \
    .head(3)
```
