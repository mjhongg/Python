```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
raw_data = pd.read_spss('Koweps_hpwc14_2019_beta2.sav')
raw_data
```

```python
welfare = raw_data.copy()
welfare
```

```python
welfare = welfare.rename( 
        columns = { 'h14_g3' : 'sex',
                    'h14_g4' : 'birth',
                    'h14_g10' : 'marriage_type',
                    'h14_g11' : 'religion',
                    'p1402_8aq1' : 'income',
                    'h14_eco9' : 'code_job',
                    'h14_reg7' : 'code_region' } )
welfare
```

```python
welfare['sex'].dtypes
```

```python
welfare['sex'].value_counts()
```

```python
welfare['sex'] = np.where( welfare['sex'] == 9, np.nan, welfare['sex'] )
welfare['sex'].isna().sum()
```

```python
welfare['sex'] = np.where( welfare['sex'] == 1, 'male', 'female' )
welfare['sex'].value_counts()
```

```python
welfare['income'].dtypes
```

```python
welfare['income'].value_counts()
```

```python
welfare['income'] = np.where( welfare['income'] == 9999, np.nan, welfare['income'] )
```

```python
welfare['income'].isna().sum()
```

```python
sex_income = welfare.dropna( subset = ['income'] ) \
    .groupby('sex', as_index = False) \
    .agg( sex_income = ('income', 'mean') )
```

```python
welfare = welfare.assign( ageg = np.where( welfare['age'] < 30, 'young',
                                 np.where( welfare['age'] <= 59, 'middle',
                                                                  'old' ) ) )
welfare['ageg']
```

```python
sns.countplot( data = welfare, x = 'ageg' )
```

```python
ageg_income = welfare.dropna(subset = ['income']) \
    .groupby('ageg', as_index = False) \
    .agg(income_mean = ('income', 'mean'))
```

```python
ageg_income
```

```python
sns.barplot( data = ageg_income, x = 'ageg' )
```
