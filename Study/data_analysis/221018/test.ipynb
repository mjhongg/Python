```python
import pandas as pd
import numpy as np


midwest_csv = pd.read_csv('midwest.csv')
midwest_csv
```

```python
midwest_csv = midwest_csv.rename(columns = {'poptotal' : 'total'})
midwest_csv = midwest_csv.rename(columns = {'popasian' : 'asian'})
midwest_csv
```

```python
midwest_csv['total']
```

```python
midwest_csv['asian']
```

```python
midwest_csv['asian_per'] = midwest_csv['asian'] / midwest_csv['total'] * 100
midwest_csv['asian_per']
```

midwest_csv

```python
midwest_csv['asian_per']
```

```python
midwest_csv['asian_per'].plot.hist()
```

```python
midwest_csv['asian_per'].mean()
```

```python
midwest_csv['per_size'] = np.where( midwest_csv['asian_per'] > 0.4872, 'large', 'small' )
midwest_csv['per_size']
```
